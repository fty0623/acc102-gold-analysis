```python
import akshare as ak
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1. Fetch historical data for the China Gold ETF (Ticker: 518880) directly from domestic servers
print("Fetching Gold ETF data from domestic financial servers (No VPN needed)...")
gold_df = ak.fund_etf_hist_em(symbol="518880", period="daily", start_date="20210419", end_date="20260419")

# 2. Data Cleaning & Transformation
# Keep essential columns and translate Chinese headers to English to meet assignment rules
gold_df = gold_df[['日期', '开盘', '收盘', '最高', '最低', '成交量']]
gold_df.columns = ['Date', 'Open', 'Close', 'High', 'Low', 'Volume']

# Convert 'Date' to datetime format and set as the index
gold_df['Date'] = pd.to_datetime(gold_df['Date'])
gold_df.set_index('Date', inplace=True)

# 3. Save to the local 'data' folder
os.makedirs('./data', exist_ok=True)
gold_df.to_csv('./data/china_gold_etf.csv')
print("Success! Data saved to './data/china_gold_etf.csv'. Generating chart...")

# 4. Data Visualization
plt.figure(figsize=(12, 6))
# Plotting the closing price trend in crimson red
plt.plot(gold_df.index, gold_df['Close'], color='crimson', linewidth=2)

# Adding user-friendly labels for the target audience (Individual Investors)
plt.title('Historical Trend of China Gold ETF (Ticker: 518880)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Closing Price (CNY)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 5. Display the final chart
plt.show()
```
